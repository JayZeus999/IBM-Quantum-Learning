```python
from qiskit import QuantumCircuit
```

```python
qc = QuantumCircuit(1)
```

```python
# X gate

qc.x(0)
qc.draw("mpl")
```

```python
# H gate

qc = QuantumCircuit(1)
qc.h(0)
qc.draw("mpl")
```

```python
# Z gate: Adds a phase shift of Δϕ = π to the ∣1⟩ state:

qc = QuantumCircuit(1)
qc.z(0)
qc.draw("mpl")
```

```python
# T gate: Adds a phase shift of Δϕ = π/4 to the ∣1⟩ state:

qc = QuantumCircuit(1)
qc.t(0)
qc.draw("mpl")
```
