```python
from qiskit import QuantumCircuit
```

```python
qc = QuantumCircuit(1)
```

```python
# X gate

qc.x(0)
qc.draw("mpl")
```

```python
# H gate

qc = QuantumCircuit(1)
qc.h(0)
qc.draw("mpl")
```

```python
# Z gate: Adds a phase shift of Δϕ = π to the ∣1⟩ state:

qc = QuantumCircuit(1)
qc.z(0)
qc.draw("mpl")
```

```python
# T gate: Adds a phase shift of Δϕ = π/4 to the ∣1⟩ state:

qc = QuantumCircuit(1)
qc.t(0)
qc.draw("mpl")
```

```python
qc = QuantumCircuit(2)
qc.cx(0, 1)
qc.draw("mpl")
```

```python
qc = QuantumCircuit(2)
qc.swap(0, 1)
qc.draw("mpl")
```

```python
# The SWAP gate can actually be constructed from three CNOTs.

qc = QuantumCircuit(2)
qc.swap(0, 1)
qc.decompose().draw("mpl")
```

Toffoli gate (or "controlled-controlled-NOT" (CCNOT)): This is a three-qubit gate

There are two control qubits and one target qubit, and the state of the target qubit is flipped only if both control qubits are in the state ∣1⟩.

```python
qc = QuantumCircuit(3)
qc.ccx(0, 1, 2)
qc.draw("mpl")
```

```python
qc = QuantumCircuit(
    1, 1
)  # the second number is the number of classical bits in the circuit
qc.measure(0, 0)
qc.draw("mpl")
```
